###Task: define a function that print whether or not given *n* is even, *n* is a parameter

```python
def inter(n):
    for i in range(1, n+1):
      prime = True
      for j in range (2,i):
        if i % j == 0:
          prime = False
          break
      if prime:
        print(i)
inter(10)
        
```

###Task: define a function that print whether or not given *n* is prime, *n* is a parameter

```python
def prime(n):
  prime = True
  for i in range(2, n):
    if n % i == 0:
      prime = False
      break
  if prime:
    print(n, 'is prime')
  else:
    print(n, 'is not prime')
prime(7)
prime(10)
```
